```python
import requests
import pandas as pd
import datetime

# URL and parameters for retrieving ratings
url = "https://shopee.com.my/api/v2/item/get_ratings"
shop_id = "125285825"
item_id = "20733279016"
limit = 1
offset = 0
reviews = []

# Set up session with necessary headers
session = requests.Session()
session.headers.update({"Cookie": "_gcl_au=xxxx; SPC_IA=-1; SPC_EC=-; SPC_F=xxxx; SPC_U=-; SPC_T_ID=xxxx; SPC_T_IV=xxxx; SPC_SI=xxxx; _ga=xxxx; _gid=xxxx; cto_lwid=xxxx; _fbp=xxxx; _hjid=xxxx; SPC_SIxxxx=xxxx"})

# Loop to retrieve ratings/reviews(maximum of 5 pages)(this one i did 50 because don't know how to set pages so i assume one page = 10 reviews)
for i in range(50): 
    # Parameters for API request
    params = {
        "itemid": item_id,
        "shopid": shop_id,
        "offset": offset,
        "limit": limit,
        "filter": "0",
        "flag": "0",
        "sort": "0",
        "append": "0",
        "before_bundle": "",
        "language": "en",
    }

    # Send request and receive response
    response = session.get(url, params=params).json()
    
    # Check for errors in the response
    if response["error"]:
        print(f"Error: {response['error']}")
        break
    # Check if there are ratings in the response
    elif response["data"]["ratings"]:
        print(offset)
        print(response["data"]["ratings"])

        # Loop through each rating in the response
        for rating in response["data"]["ratings"]:
            # Convert Unix timestamp to datetime object and format the datetime
            rating_time = datetime.datetime.fromtimestamp(rating['ctime'])
            rating['ctime'] = rating_time.strftime('%Y-%m-%d %H:%M:%S')  # Convert to string in desired format
            
            # Extract 'model_name' from 'product_items' column and store it in the 'model_name' column of the DataFrame 'df'
            model_name = rating.get('product_items')[0].get('model_name')
            rating['model_name'] = model_name
            
            # Append the rating to the list of reviews
            reviews.append(rating)
            
            # Increment the offset for the next page of ratings
            offset += limit

            # Add a delay between requests (optional)
            # time.sleep(1)

    else:
        # If there are no more ratings, exit the loop
        break

# Create a DataFrame from the list of reviews
df = pd.DataFrame(reviews)

print(f"Retrieved {len(reviews)} reviews.")

# Select columns to keep in the final CSV file (only include reviewer name, review date, rating, and comment)
df = df[['author_username', 'ctime', 'rating_star', 'comment']]

print(df.columns)

# Save the DataFrame as a CSV file
df.to_csv('Reviews.csv', index=False)
```
